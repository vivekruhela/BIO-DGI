```python
ghis_score = pd.read_excel("/home/vivek/jupyter_notebooks/bio_dgi_extension/final_compilation_work/haploinsufficiency/GHIS_scores.xlsx", index_col=0)
ghis_score.head()
```

```python
df = pd.DataFrame()
comm, katz, kvhigh, ghis = [], [], [], []
present_in_gene282 = []
ghis_vhigh = []
for gen in tqdm(top500genes):
    if gen in gene282_data1["elements"].values.tolist():
        present_in_gene282.append(1)
    else:
        present_in_gene282.append(0)
        
    try:
        comm.append(gene_katz.loc[gen, "community"])
        katz.append(gene_katz.loc[gen, "katz_centrality"])
        if comm[-1] == "C1" and katz[-1] >= 0.03664322192980843:
            if katz[-1] >= 0.05918131877667836:
                kvhigh.append(2)
            else:
                kvhigh.append(1)
        elif comm[-1] == "C2" and katz[-1] >= 0.04247963851982776:
            if katz[-1] >= 0.08188751918525802:
                kvhigh.append(2)
            else:
                kvhigh.append(1)
        elif comm[-1] == "C3" and katz[-1] >= 0.06212123517731534:
            if katz[-1] >= 0.09493436620191388:
                kvhigh.append(2)
            else:
                kvhigh.append(1)
        elif comm[-1] == "C4" and katz[-1] >= 0.023394482986119285:
            if katz[-1] >= 0.05256923673545837:
                kvhigh.append(2)
            else:
                kvhigh.append(1)
        elif comm[-1] == "C5" and katz[-1] >= 0.003945845861411295:
            if katz[-1] >= 0.02337439872602132:
                kvhigh.append(2)
            else:
                kvhigh.append(1)
        else:
            kvhigh.append(0)
    except:
        comm.append(np.nan)
        katz.append(np.nan)
        kvhigh.append(np.nan)
        
    try:
        ghis.append(ghis_score[ghis_score["name"] == gen]["GHIS"].values[0])
        if ghis[-1] >= 0.52:
            if ghis[-1] >= 0.59:
                ghis_vhigh.append(2)
            else:
                ghis_vhigh.append(1)
        else:
            ghis_vhigh.append(0)
    except:
        ghis.append(np.nan)
        ghis_vhigh.append(np.nan)
        
df["gene"] = top500genes
df["282panel"] = present_in_gene282
df["community"] = comm
df["katz"] = katz
df["kvhigh"] = kvhigh
df["ghis"] = ghis
df["ghis_vhigh"] = ghis_vhigh
df = df.fillna(0)
df.to_excel("gene500_katz_haplo.xlsx", index=False)
df.head()
```

```python
df1 = df[df["community"] == "C5"]
df1 = df1[["kvhigh","ghis_vhigh"]]
corr_matrix = df1.corr(method="spearman")
corr_matrix.style.background_gradient(cmap='coolwarm')
```
